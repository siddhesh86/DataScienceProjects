```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
import scipy.stats as sts

```

```python
def make_data(N, f=0.3, rseed=1):
    rand = np.random.RandomState(rseed)
    x = rand.randn(N)
    print(f"{int(f * N) = }")
    print(f"\n\nx befpre: {x}\n\n")
    x[int(f * N):] += 5
    return x

x = make_data(1000)
x
```
