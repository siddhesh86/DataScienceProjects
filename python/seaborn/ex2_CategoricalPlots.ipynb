```python
import seaborn as sns
import numpy as np

data=sns.load_dataset('tips')
data
```

```python
## categorical plots
sns.catplot(data=data, x='day', y='total_bill', hue="smoker", kind='box')
```

```python
sns.boxplot(data=data, x='day', y='total_bill', hue="smoker")
```

```python
sns.catplot(data=data, x='day', y='total_bill', hue="smoker", kind='violin')
```

```python
sns.violinplot(data=data, x='day', y='total_bill', hue="smoker", split=True)
```

```python
sns.catplot(data=data, x='day', y='total_bill', hue="smoker", kind='bar')
```

```python
sns.barplot(data=data, x='day', y='total_bill', hue="smoker", estimator=np.mean)
```

```python
sns.catplot(data=data, x='day', hue="smoker", kind='count')
```

```python
sns.countplot(data=data, x='day', hue="smoker")
```

```python
sns.catplot(data=data, x='day', y='total_bill', hue="smoker", kind='strip')
```

```python
sns.stripplot(data=data, x='day', y='total_bill', hue="smoker")
```

```python
sns.catplot(data=data, x='day', y='total_bill', hue="smoker", kind='swarm')
```

```python
sns.swarmplot(data=data, x='day', y='total_bill', hue="smoker")
```

```python
sns.catplot(data=data, x='day', y='total_bill', hue="smoker", kind='boxen')
```
