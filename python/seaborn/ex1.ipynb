```python
import seaborn as sns

data = sns.load_dataset("tips")
data
```

```python
sns.distplot(data['total_bill'])
```

```python
sns.displot(data=data,x="total_bill")
```

```python
sns.displot(data=data,x="total_bill", y='tip')
```

```python
fig = sns.histplot(data=data, x="total_bill")
fig.set_xlim(0, 50)
fig.set_title('my title')
```

```python
sns.histplot(data=data, x="total_bill", y='tip')
```

```python
sns.catplot(data=data, x='day',y='total_bill', hue='sex', kind='swarm')
```

```python
sns.catplot(data=data, x='day',y='total_bill', hue='sex', kind='violin')
```

```python
sns.catplot(data=data, x='day',y='total_bill', hue='sex', kind='bar')
```

```python
sns.jointplot(data=data, x='total_bill', y='tip', hue='sex')
```

```python
sns.pairplot(data=data, hue='sex')
```

```python
sns.pairplot(data=data, hue='sex', kind='hist')
```

```python
g = sns.PairGrid(data=data, hue='sex')
#g.map(sns.scatterplot)
# or
#g.map_diag(sns.displot)
g.map_diag(sns.histplot)
#g.map_offdiag(sns.scatter)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```
