```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor 
from sklearn.metrics import mean_absolute_error

train_full = pd.read_csv("home-data-for-ml-course/train.csv", index_col="Id")
test_full  = pd.read_csv("home-data-for-ml-course/test.csv", index_col="Id")

print("train_full.shape: {}, train_full.columns: {}".format(train_full.shape, train_full.columns))
print("\ntest_full.shape: {}, test_full.columns: {}".format(test_full.shape, test_full.columns))
print("\ntrain_full.describe():\n {}".format(train_full.describe()))      
```

```python
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']

X = train_full[features].copy()
y = train_full['SalePrice']

X_test = test_full[features].copy()

print("X: {}".format(X))
print("\ny:{}".format(y))
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=0)

print("X_train: {}".format(X_train))
print("\nX_val: {}".format(X_val))

print("\ny_train: {}".format(y_train))
print("\ny_val: {}".format(y_val))
```

```python
model_RandomForestRegressor = RandomForestRegressor(n_estimators=100, random_state=0, criterion='absolute_error') 
model_RandomForestRegressor.fit(X_train, y_train)
print("mean_absolute_error(validation): {}".format(mean_absolute_error(y_val, model_RandomForestRegressor.predict(X_val))))
```

```python
y_test = model_RandomForestRegressor.predict(X_test)

result = pd.DataFrame({"ID": X_test.index, "SalePrice": y_test})
result.to_csv("submission.csv", index=False)
print("result: {}".format(result))
```
