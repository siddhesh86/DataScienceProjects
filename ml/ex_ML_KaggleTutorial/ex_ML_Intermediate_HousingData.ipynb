```python
print("dataset.isna(): \n{}".format(dataset.isna()))
```

```python
print("\ndataset.isna().any(): \n{}".format(dataset.isna().any()))
```

```python
y = dataset['Price']

dataset_predictor = dataset.drop(['Price'], axis='columns')
X = dataset_predictor.select_dtypes(exclude='object')
X.describe()
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=0)
```
