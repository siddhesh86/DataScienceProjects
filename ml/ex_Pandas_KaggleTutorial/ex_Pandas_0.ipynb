```python
################################################################################
### Maps-------
###

review_points_mean = reviews.points.mean()
print("review_points_mean:", review_points_mean)
```

```python
review_points_map = reviews.points.map(lambda p: p - review_points_mean)
reviews.points.describe()
```

```python
print("review_points_mean:", review_points_mean)
def remean_points(row):
    row.points = row.points - review_points_mean
    return row

reviews_clone1 = reviews.copy()
reviews_clone2 = reviews_clone1.apply(remean_points, axis='columns')
```

```python
reviews['points_to_price'] = reviews.points / reviews.price

reviews.describe()
reviews.head(5)
```

```python
bargain_wine1 = reviews.loc[(reviews.price != 'Nan') & (reviews.points_to_price == reviews.points_to_price.max()), 'title']
bargain_wine = bargain_wine1.iloc[0]
print("bargain_wine1: {}".format(bargain_wine1))
print("bargain_wine: {}".format(bargain_wine))
```

```python
desc_topical = reviews.description.map(lambda desc: "tropical" in desc)
n_topical = desc_topical.sum()

desc_fruity = reviews.description.map(lambda desc: "fruity" in desc)
n_fruity = desc_fruity.sum()

descriptor_couts = pd.Series([n_topical, n_fruity], index=['topical', 'fruity'])

print("desc_topical: {}".format(desc_topical))
print("desc_topical.sum: {}".format(desc_topical.sum()))

print("descriptor_couts: {}".format(descriptor_couts))
```

```python
def allocate_star(row):
    row.star_ratings = 1
    if row.points >= 95:
        row.star_ratings = 3
    elif row.points >= 85:
        row.star_ratings = 2
        
    if row.country == 'Canada':
        row.star_ratings = 3
        
    return row;

reviews['star_ratings'] = 1
reviews_11 = reviews.apply(allocate_star, axis=1)
star_ratings = reviews_11['star_ratings']

print("star_ratings: {}".format(star_ratings))  
print("reviews.star_ratings.value_counts(): {}".format(reviews.star_ratings.value_counts()))
print("reviews_11: {}".format(reviews_11))        
    
```
