```python
features = ['LotArea', 'OverallQual', 'YearBuilt', 'GrLivArea', '1stFlrSF', '2ndFlrSF', 'BedroomAbvGr', 'TotRmsAbvGrd', 'FullBath', 'GarageArea' ]
X = home_Iowa_data[features]
y = home_Iowa_data.SalePrice

from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X, y)
```

```python
# DecisionTreeRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error

random_state_1 = 1

def DecisionTree(train_X, val_X, train_y, val_y, max_leaf_nodes):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=random_state_1)
    model.fit(train_X, train_y)
    return mean_absolute_error(val_y, model.predict(val_X))

mae_min = 1e+10
max_leaf_nodes_best = None
for max_leaf_nodes in [10, 25, 50, 100, 500, 1000, 5000]:
    mae = DecisionTree(train_X, val_X, train_y, val_y, max_leaf_nodes)
    if mae < mae_min:
        max_leaf_nodes_best = max_leaf_nodes
        mae_min = mae
        
model_DessionTreeRegressor = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes_best, random_state=random_state_1)
model_DessionTreeRegressor.fit(train_X, train_y)
print("DecisionTree: error: {}".format(mean_absolute_error(val_y, model_DessionTreeRegressor.predict(val_X))))
```

```python
from sklearn.ensemble import RandomForestRegressor

model_RandomForestRegressor = RandomForestRegressor(random_state=random_state_1)
model_RandomForestRegressor.fit(train_X, train_y)
print("RandomForest: error: {}".format(mean_absolute_error(val_y, model_RandomForestRegressor.predict(val_X))))
```
